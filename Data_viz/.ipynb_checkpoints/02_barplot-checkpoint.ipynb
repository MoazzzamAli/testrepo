# BarPlot Day 2

```python
import seaborn as sns
import numpy
import matplotlib.pyplot as plt
sns.set_theme(style="ticks", color_codes=True)
#import data_set
phool = sns.load_dataset("iris")
#phool=sns.load_dataset("healthexp")

# draw a barplot
sns.barplot(x="species", y="sepal_width", data=phool)
plt.show()
```

```python
phool
```

```python
sns.barplot(x="species", y="petal_length", data=phool)
plt.show()
```

```python
# loading titanic dataset
kashti=sns.load_dataset("titanic")
kashti
```

```python
sns.barplot(x="who",y="alone", hue="sex", data=kashti)
```

```python
sns.barplot(x="sex",y="alone", hue="who", data=kashti)
```

```python
# atribute "who" adds the legends to barcharts
sns.barplot(x="who",y="alone", hue=None, data=kashti)
```

```python
# how to set the order of columns
sns.barplot(x="sex",y="alone", hue="who", data=kashti, order=["female", "male"])
```

```python
# Change the color of bars
sns.barplot(x="sex",y="alone", hue="who", data=kashti, order=["female", "male"], color="grey")
```

```python
sns.barplot(x="sex",y="alone", hue="who", data=kashti, order=["female", "male"], color="red", ci=None)
```

```python
sns.barplot
```

```python
sns.barplot(x="sex",y="alone", hue="who", data=kashti, order=["female", "male"], color="red", errorbar=None,palette="pastel")
```

```python

#using estimators
sns.barplot(x="sex",y="alone", hue="who", data=kashti, estimator=numpy.median)
```
