# Adding Limits

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="ticks", color_codes=True)
#import data_set
phool = sns.load_dataset("iris")
#phool=sns.load_dataset("healthexp")

# lineplot
sns.lineplot(x="sepal_length", y="sepal_width", data=phool)
plt.title("Data of Flowers")
plt.xlim(4)
plt.ylim(3)
```

## size of Figure

```python
import seaborn as sns
import matplotlib.pyplot as plt

#import data_set
phool = sns.load_dataset("iris")
#phool=sns.load_dataset("healthexp")
plt.figure(figsize=(10,8))
# lineplot
sns.lineplot(x="sepal_length", y="sepal_width", data=phool)

plt.title("Data of Flowers")

sns.set_theme(style="dark", palette='deep', color_codes=True)
```
