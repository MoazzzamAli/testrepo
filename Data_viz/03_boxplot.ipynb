# Adding individual Palette color

```python
import seaborn as sns

sns.set(style="darkgrid")
tip=sns.load_dataset("tips")
# adding palette color
mypalette = {
    'Yes': 'tab:blue',
    'No': 'tab:green',
    }

sns.boxplot(x="tip", y="day",data=tip,saturation=1, hue="smoker", palette=mypalette)
```

```python

p1=sns.boxplot(x="survived",
           y="age", showmeans=True,
           data=kashti)
```

```python
import matplotlib.pyplot as plt
p1=sns.boxplot(x="survived",
           y="age", showmeans=True,
           meanprops={"marker": "+", "markersize":"12","markeredgecolor":"red"},
               data=kashti)
plt.xlabel("How many Survived", size=10)
plt.ylabel("Age (years)", size=10)
plt.title("Box plot of kitnay doobay or kitnay bach gaye", size=16,weight="bold")
plt.show()
```
